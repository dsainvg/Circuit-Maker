```python
import pandas as pd
df = pd.DataFrame({'A3':[], 'A2':[], 'A1':[], 'A0':[],'B3':[], 'B2':[], 'B1':[], 'B0':[]})
df_o = pd.DataFrame({'CO':[],'C3':[], 'C2':[], 'C1':[], 'C0':[]})
```

```python
for i in range(16):
    for j in range(16):
        a = [int(x) for x in f"{i:04b}"]
        b = [int(x) for x in f"{j:04b}"]
        c = [int(x) for x in f"{(i+j):05b}"]
        print(a, b, c)
        df.loc[len(df)] = a + b
        df_o.loc[len(df_o)] = c
        
```

```python

df_o.to_csv('output.csv', index=False)
df.to_csv('input.csv', index=False)
```
