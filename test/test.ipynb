```python
import pandas as pd
df = pd.DataFrame({'T3':[], 'T2':[], 'T1':[], 'T0':[],'O3':[], 'O2':[], 'O1':[], 'O0':[]})
```

```python
for i in range(64):
    ones = i % 10
    tens = (i // 10)
    ones_str = list(bin(ones)[2:].zfill(4))
    tens_str = list(bin(tens)[2:].zfill(4))
    df.loc[i] = tens_str + ones_str    
df.to_csv('test.csv', index=False) 
    
```
