```python
import pandas as pd
df = pd.DataFrame({'A1':[], 'A0':[],'B1':[], 'B0':[]})
df_o = pd.DataFrame({'CO':[],'C1':[], 'C0':[]})
```

```python
for i in range(4):
    for j in range(4):
        a = [int(x) for x in f"{i:02b}"]
        b = [int(x) for x in f"{j:02b}"]
        c = [int(x) for x in f"{(i+j):03b}"]
        print(a, b, c)
        df.loc[len(df)] = a + b
        df_o.loc[len(df_o)] = c
        
```

```python

df_o.to_csv('output.csv', index=False)
df.to_csv('input.csv', index=False)
```
